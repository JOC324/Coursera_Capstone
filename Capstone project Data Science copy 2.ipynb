```python
# Table keeps the information that is found in the wikipedia web page
table=pd.read_html(link)
```

```python
# Eliminating the row where the value of Borough is "Not assigned"
df_canada=df_canada.drop(df_canada[df_canada.Borough=='Not assigned'].index)
```

```python
# Updating the index of new data frame
df_canada.reset_index(drop=True, inplace=True)
```

```python
path='http://cocl.us/Geospatial_data'
```

```python
df_coord=pd.read_csv(path)
```

```python
df_coord.head()
```

```python
df_canada.iloc[0,0]
    
```

```python
list_lat=[]
list_long=[]
list_data=[]
for data in df_canada['Postal Code']:
    lat=0
    long=0 
    n_row=0
    for data2 in df_coord['Postal Code']:
        if(data==data2):
            lat=df_coord.loc[n_row,'Latitude']
            long=df_coord.loc[n_row,'Longitude']
            list_data.append(data)
            list_lat.append(lat)
            list_long.append(long)
        else:
            n_row=n_row+1

        
       
            
            
```

```python
len(list_lat)
```

```python
n_row
```

```python
list_long
```

```python
coord=[list_data,list_lat, list_long]
```

```python
df_coord2=pd.DataFrame(coord)
```

```python
df_coord2
```

```python
df_coord3=df_coord2.transpose()
```

```python
df_coord3.head()
```

```python
df_coord3.rename(columns={0:'Postal Code',1:'Latitude',2:'Longitude'}, inplace=True)
```

```python
df_coord3
```

```python
pd_canada_final=pd.merge(df_canada, df_coord3, on='Postal Code')
```
